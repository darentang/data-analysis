```python
import pandas as pd

therapeutic = pd.read_csv("data/Function Therapeutic result.txt", sep="\t")
placebo = pd.read_csv("data/Function Placebo result.txt", sep="\t")
therapeutic["name"] = therapeutic.name.str.rstrip()
placebo["name"] = placebo.name.str.rstrip()
placebo["subgroup"] = placebo.subgroup.str.replace(" Placebo", "")
therapeutic["subgroup"] = therapeutic.subgroup.str.replace(" Therapeutic", "")

placebo
```

```python
therapeutic["name"] = [" ".join(x.split()[:2]) for x in therapeutic["name"]]
placebo["name"] = [" ".join(x.split()[:2]) for x in placebo["name"]]
```

```python
therapeutic = therapeutic.set_index("name").sort_index()
placebo = placebo.set_index("name").sort_index()
```

```python
sizes = pd.read_csv("sizes.txt", sep="\t")
sizes["group"] = ["placebo" if "Placebo" in x else "therapeutic" for x in sizes.subgroup]
sizes["name"] = [" ".join(x.split()[:2]) for x in sizes.name]
sizes = sizes.set_index("name").sort_index()
sizes
```

```python
sizes = sizes.loc[therapeutic.index, :]
```

```python
assert all(therapeutic.index == placebo.index) 
```

```python
assert all(sizes.query("group == 'therapeutic'").index == therapeutic.index)
assert all(sizes.query("group == 'placebo'").index == placebo.index)
```

```python
placebo["n"] = sizes.query("group == 'placebo'").pre
therapeutic["n"] = sizes.query("group == 'therapeutic'").pre

placebo
```

```python
therapeutic
```

```python
print("Excluding these papers:\n")
exclusions = []
for name1, name2, mu1, mu2 in zip(placebo.index, therapeutic.index, placebo.mu, therapeutic.mu):
    assert name1 == name2
    # only exclude if both negative
    if mu1 < 0 and mu2 < 0:
        print(name1, mu1, mu2)
        exclusions.append(name1)
```

```python
import numpy as np
data = pd.DataFrame()

# R = X / Y
# clip between 1% and 100%
data["L"] = np.log(np.clip(placebo.mu / therapeutic.mu, 0.01, 1))
# Equation 1
data["v"] = placebo.sigma/placebo.n/placebo.mu**2 + therapeutic.sigma/therapeutic.n/therapeutic.mu**2
# w_i = 1 / v_i
data["w"] = 1 / data.v
data["subgroup"] = placebo.subgroup
data["df"] = placebo.n + therapeutic.n - 2
data = data.dropna()
data = data[~data.index.isin(exclusions)]
data
```

```python
from scipy.stats import norm

def calc_ratio(data):
    z_alpha = - norm.ppf(0.05/2)
    k = len(data)
    # Reject intervention subgroups k < 3
    if k < 3:
        return 0, 0, 0
    # Equation 4
    Q = np.sum(data.w * data.L**2) - np.sum(data.w*data.L)**2/np.sum(data.w)
    # Variance of lambda, Equation 3
    if k > 1:
        var_lambda = (Q - (k-1) )/(np.sum(data.w) - np.sum(data.w**2)/np.sum(data.w))
    else:
        var_lambda = 0
        
    # w_star_i = 1 / (v_i + var_lambda)
    w_star = 1/(data.v + var_lambda)
    # Equation 5
    L_star = np.sum(w_star * data.L) / np.sum(w_star)
    # Equation 7
    w_star_sum = np.sum(w_star)
    se_L = np.sqrt(1/w_star_sum * \
                   (1 + 4*np.sum(1/data.df*(w_star/data.w)**2*\
                                 (w_star*(w_star_sum-w_star))/w_star_sum**2)))
    # Approx. 1.96
    z_alpha = - norm.ppf(0.05/2)
    # Lower and upper bound log ratio
    lb = L_star - z_alpha * se_L
    ub = L_star + z_alpha * se_L

    # Exponentiating 
    mu_rho = np.exp(L_star)
    lb_rho = np.exp(lb)
    ub_rho = np.exp(ub)

    return mu_rho, lb_rho, ub_rho
```

```python
def format_result(group, mu, lb, ub):
    if mu == 0:
        print(f"Subgroup analysis cannot be done for {group}.")
        return
    print(f"{group}: {mu:.2f} [{lb:.2f}, {ub:.2f}]")


results = data.groupby("subgroup").apply(calc_ratio)

for group, (mu, lb, ub) in results.iteritems():
    format_result(group, mu, lb, ub)

mu, lb, ub = calc_ratio(data)
format_result("Overall", mu, lb, ub)
```
