```python
import pandas as pd

therapeutic = pd.read_csv("data/Pain Therapeutic result.txt", sep="\t")
placebo = pd.read_csv("data/Pain Placebo result.txt", sep="\t")
therapeutic["name"] = therapeutic.name.str.rstrip()
placebo["name"] = placebo.name.str.rstrip()
therapeutic = therapeutic.set_index("name").sort_index()
placebo = placebo.set_index("name").sort_index()
```

```python
sizes = pd.read_csv("sizes.txt", sep="\t")
sizes["group"] = ["placebo" if "Placebo" in x else "therapeutic" for x in sizes.subgroup]
sizes["name"] = [" ".join(x.split()[:2]) for x in sizes.name]
sizes = sizes.set_index("name").sort_index()
sizes
```

```python
assert all(sizes.query("group == 'therapeutic'").index == therapeutic.index)
```

```python
placebo["n"] = sizes.query("group == 'placebo'").pos
therapeutic["n"] = sizes.query("group == 'therapeutic'").pos

placebo
```

```python
therapeutic
```

```python
print("Excluding these papers:\n")
exclusions = []
for name1, name2, mu1, mu2 in zip(placebo.index, therapeutic.index, placebo.mu, therapeutic.mu):
    assert name1 == name2
    # only exclude those with double negatives
    if mu1 < 0 and mu2 < 0:
        # negating mu1
        print(name1, mu1, mu2)
        # exclusions.append(name1)
        placebo.loc[name1, "mu"] = -mu1
```

```python
import numpy as np
data = pd.DataFrame()

# clip between 1% and 100%
data["L"] = np.log(np.clip(placebo.mu / therapeutic.mu, 0.01, 1))
data["v"] = placebo.sigma/placebo.n/placebo.mu**2 + therapeutic.sigma/therapeutic.n/therapeutic.mu**2
data["w"] = 1 / data.v
data["subgroup"] = placebo.subgroup
data = data.dropna()
data = data[~data.index.isin(exclusions)]
data
```

```python
from scipy.stats import norm

def calc_ratio(data):
    k = len(data)
    if k < 3:
        return 0, 0, 0
    Q = np.sum(data.w * data.L**2) - np.sum(data.w*data.L)**2/np.sum(data.w)
    var_lambda = (Q - (k-1) )/(np.sum(data.w) - np.sum(data.w**2)/np.sum(data.w))

    w_star = 1/(data.v + var_lambda)
    mu_lambda = np.sum(w_star * data.L) / np.sum(w_star)
    se_L = np.sqrt(1 / np.sum(w_star))

    z_alpha = - norm.ppf(0.05/2)
    lb = mu_lambda - z_alpha * se_L
    ub = mu_lambda + z_alpha * se_L

    mu_rho = np.exp(mu_lambda)
    lb_rho = np.exp(lb)
    ub_rho = np.exp(ub)

    return mu_rho, lb_rho, ub_rho
```

```python
def format_result(group, mu, lb, ub):
    if mu == 0:
        print(f"{group} is too small.")
        return
    print(f"{group}: {mu:.2f} [{lb:.2f}, {ub:.2f}]")


results = data.groupby("subgroup").apply(calc_ratio)

for group, (mu, lb, ub) in results.iteritems():
    format_result(group, mu, lb, ub)

mu, lb, ub = calc_ratio(data)
format_result("Overall", mu, lb, ub)
```
